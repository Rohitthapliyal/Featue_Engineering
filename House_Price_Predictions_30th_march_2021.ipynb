```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#display all the columns of the datafarme
pd.pandas .set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv(r'C:\Users\rohit\Downloads\house-prices-advanced-regression-techniques\train.csv')
dataset.head()
```

```python
print(dataset.shape)
```

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4),'%missing values')
```

```python
#realtionship between the missing values and the sales price

for feature in features_with_na:
    data=dataset.copy()
    
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar(color=['red','blue'])
    plt.title(feature)
    plt.show()
```

```python
#numerical features

numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes !='O']
print('number of numerical variables :',len(numerical_features))

dataset[numerical_features].head()
```

```python
#Temporal variable eg-datetime variables

year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
#analyze temporal datetime variables
#check whether there is realtion between year  the house is sold and the sales price

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('YearSold')
plt.ylabel('Median house price')
plt.title('House Price vs YearSold')
```

```python
year_feature
```

```python
#compare difference between all year features with sales price

for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

# Feature engineering

```python
#logarithmic transformation

for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title('feature')
        plt.show()
        
```

```python
#outliers

for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
